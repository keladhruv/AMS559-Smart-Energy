```python
import numpy
import matplotlib.pyplot as plt
```

```python
y=numpy.loadtxt('TestData.txt')
```

```python
def gradient(x,y,index):
    slope = 0.15
    if y[index]<=x[index]:
        if x[index]<x[index-1]:
            slope-=1
        elif x[index]>x[index-1]:
            slope+=1
    elif y[index]>x[index] and x[index]<x[index-1]: 
        slope -=2
    return slope
```

```python
def gradientDesOptimization(y,steps=0.2):
    n=len(y)
    x=[0.0]*(n+1)
    x[0]=y[0]
    for i in range(n):
        x[i+1]=x[i]- steps*gradient(x,y,i)
    outX = x[1:]
    return outX
```
