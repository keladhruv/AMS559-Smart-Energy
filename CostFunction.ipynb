```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import numpy
```

```python
x1=pd.read_csv('outputOD.txt')
y1=pd.read_csv('TestData.txt')
x=numpy.loadtxt('outputOD.txt')
y=numpy.loadtxt('TestData.txt')
```

```python
len(x)
len(y)
```

```python
def cost(s,x,y):
    val=0
    p=0.4
    a=4
    b=4
    for i in range(1,len(x)):
        val=val +p*x[i] + a*max(0,y[i]-x[i]+ b*abs(x[i]-x[i-1]))                         
    print s,val
cost("Offline House 1",x,y)
```

```python
plt.title("Optimized Values vs Actual Values (Offline)")
plt.plot(x, label="Optimized Values")
plt.plot(y, label="Actual Values")
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.ylabel("Electricity Units")
plt.xlabel("Time (1Unit=15Mins)")
```
